Finalmente sedecide transformar los datos de la siguiente manera:
- 1 al 19: 1
- 20 al 29: 2
- 30 al 39: 3

```python
df['EnvironmentSatisfaction'] = np.where(
    df['EnvironmentSatisfaction'] <= 19.99,
    1.0,
    np.where(
        df['EnvironmentSatisfaction'] <= 29.99,
        2.0,
        np.where(
            df['EnvironmentSatisfaction'] <= 39.99,
            3.0,
            4.0
        )
    )
)
```

## Gestión de nulos

```python
(df.isnull().sum()[df.isnull().sum() > 0]/df.shape[0] * 100).round(2)
```

Se podrian imputar por su bajo porcentaje de nulos:
- DailyRate
- HourlyRate
- WorkLifeBalance  

MonthlyIncome: al faltarnos información sobre qué es esa columna, tener tantos nulos que tampoco podemos trabajar con ella y teniendo otros indicadores economicos que si podemos usar , se decide borrar esta fila.  
Mientras, los nulos que faltan al ser un porcentaje tan alto se decide cambiar a "no-data"



```python
df.drop(columns= 'MonthlyIncome',inplace= True)
```

DailyRate

```python
df['DailyRate'].describe()
```

Decidimos hacer una imputación en función del JobRole y JobLevel

```python
df.groupby(["JobRole", "JobLevel"])["DailyRate"].mean()
```

```python
moda_dailyrate = round(df.groupby(["JobRole","JobLevel"])["DailyRate"].apply(lambda x: x.mode()[0]), 2)
```

```python
# Crear una función para reemplazar los valores nulos con la moda correspondiente
def reemplazar_moda(row):
    if pd.isnull(row['DailyRate']):
        return moda_dailyrate.loc[row['JobRole'], row['JobLevel']]
    else:
        return row['DailyRate']

# Aplicar la función fila por fila
df['DailyRate'] = df.apply(reemplazar_moda, axis=1)

df["DailyRate"].isna().sum()
```

HourlyRate

```python
moda_hourly = round(df.groupby(["JobRole","JobLevel"])["HourlyRate"].apply(lambda x: x.mode()[0]), 2)
```

```python
# Crear una función para reemplazar los valores nulos con la moda correspondiente
def reemplazar_moda(row):
    if pd.isnull(row['HourlyRate']):
        return moda_hourly.loc[row['JobRole'], row['JobLevel']]
    else:
        return row['HourlyRate']

# Aplicar la función fila por fila
df['HourlyRate'] = df.apply(reemplazar_moda, axis=1)

df["HourlyRate"].isna().sum()
```

PerformanceRating

```python
moda_performance = round(df.groupby(["JobRole","JobLevel"])["PerformanceRating"].apply(lambda x: x.mode()[0]), 2)
```

```python
# Crear una función para reemplazar los valores nulos con la moda correspondiente
def reemplazar_moda(row):
    if pd.isnull(row['PerformanceRating']):
        return moda_performance.loc[row['JobRole'], row['JobLevel']]
    else:
        return row['PerformanceRating']

# Aplicar la función fila por fila
df['PerformanceRating'] = df.apply(reemplazar_moda, axis=1)

df["PerformanceRating"].isna().sum()
```
